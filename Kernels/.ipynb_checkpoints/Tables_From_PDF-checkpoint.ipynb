```python
iterables = [
    ['Getting_Gas_befor_War', 'Getting_Gas_in_War'],
    ['Bad', 'Medium', 'Good', 'Unknown']
]
columns_eng_0 = pd.MultiIndex.from_product(iterables, names=['Time', 'Degree'])
```

```python
rows_num = len(new_index_eng)
cols_num = columns_eng_0.shape[0]

table_values = np.zeros((rows_num, cols_num))

new_table_0 = pd.DataFrame(table_values, index=new_index_eng, columns=columns_eng_0)
```

```python
new_table_0[('Getting_Gas_befor_War', 'Medium')] = \
                        table_0[table_0.columns[1]].iloc[3:-1].apply(lambda x : x.split(' ')[2]).values

new_table_0[('Getting_Gas_befor_War', 'Good')] = \
                        table_0[table_0.columns[1]].iloc[3:-1].apply(lambda x : x.split(' ')[1]).values

new_table_0[('Getting_Gas_in_War', 'Bad')] = \
                        table_0[table_0.columns[0]].iloc[3:-1].apply(lambda x : x.split(' ')[3]).values

new_table_0[('Getting_Gas_in_War', 'Medium')] = \
                        table_0[table_0.columns[0]].iloc[3:-1].apply(lambda x : x.split(' ')[2]).values

new_table_0[('Getting_Gas_in_War', 'Good')] = \
                        table_0[table_0.columns[0]].iloc[3:-1].apply(lambda x : x.split(' ')[1]).values

new_table_0[('Getting_Gas_in_War', 'Unknown')] = \
                        table_0[table_0.columns[0]].iloc[3:-1].apply(lambda x : x.split(' ')[0]).values
```
